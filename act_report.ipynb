```python
plt.scatter(Last_frame['favorite_count'], Last_frame['retweet_count'], label = "values" )

plt.ylabel('retweet_count')
plt.xlabel('favorite_count')

plt.title('retweet_count Vs favorite_count')
plt.legend()
plt.show()
```
